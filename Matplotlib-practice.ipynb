```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(-4*np.pi, 4*np.pi, 2000)
```

```python
y = np.sinh(x)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 0.5, 1])
axes2 = fig.add_axes([0.5, 0, 0.5, 1])
axes1.plot(x, y, ".")
axes2.plot(y, x, "*")
```
